```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set_style('whitegrid')
```

```python
dataset = pd.read_csv('multisat.csv')
```

```python
dataset
```

```python
plt.scatter(dataset['SAT'],dataset['GPA'])
plt.show()
```

```python
plt.scatter(dataset['Rand 1,2,3'],dataset['GPA'])
plt.show()
```

```python
x = dataset[['SAT','Rand 1,2,3']]
```

```python
x
```

```python
y = dataset['GPA']
```

```python
y
```

```python
import statsmodels.api as sm
```

```python
x1 = sm.add_constant(x)
```

```python
x1
```

```python
model = sm.OLS(y,x1).fit()
```

```python
model.summary()
```

```python
test = pd.DataFrame([[1800,2],[2000,1]],columns=['SAT','Rand 1,2,3'])
```

```python
test
```

```python
test = sm.add_constant(test)
```

```python
test
```

```python
model.predict(test)
```

```python
plt.scatter(dataset['SAT'],dataset['GPA'])
plt.plot(dataset['SAT'],model.predict(x1),color='green')
plt.show
```
